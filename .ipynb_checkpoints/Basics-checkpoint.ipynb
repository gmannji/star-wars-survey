# Analyzing data on Star Wars movies

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.columns.tolist()
```

```python
star_wars.head(10)
```

```python
#Filter the data frame so it consists only rows where RespondendID is not null
star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].head(10))
yes_no = {
    "Yes": True,
    "No": False
}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].head(10)
```

```python
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts())
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].head(10))
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].head(10)
```

```python
star_wars['Which of the following Star Wars films have you seen? Please select all that apply.']
print(star_wars['Which of the following Star Wars films have you seen? Please select all that apply.'].value_counts())
print(star_wars['Unnamed: 8'].value_counts())
#print(star_wars['Unnamed: 4'].head())
```

```python
from numpy import nan

title_true = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones' : True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    nan: False
}

#star_wars['Which of the following Star Wars films have you seen? Please select all that apply.'].map(title_true_false)
#abc = star_wars.columns[3:9].map(title_true)

for i in star_wars.columns[3:9]:
    #print(star_wars[i])
    star_wars[i] = star_wars[i].map(title_true)

#star_wars[star_wars.columns[3:9]] = star_wars[star_wars.columns[3:9]].map(title_true)

print(star_wars[star_wars.columns[3:9]].head())
```

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    'Unnamed: 4': 'seen_2',
    'Unnamed: 5': 'seen_3',
    'Unnamed: 6': 'seen_4',
    'Unnamed: 7': 'seen_5',
    'Unnamed: 8': 'seen_6'
})

seen = star_wars.columns[3:9]

print(seen)
```

```python
print(star_wars.columns[9:15])
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].iloc[1:].astype(float)# = star_wars[star_wars.columns[9:15]].astype(float)
print(star_wars[star_wars.columns[9:15]].head())
```

```python
star_wars = star_wars.rename(columns={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
        'Unnamed: 10':'ranking_2', 
        'Unnamed: 11':'ranking_3', 
        'Unnamed: 12':'ranking_4', 
        'Unnamed: 13':'ranking_5',
        'Unnamed: 14':'ranking_6'
       })

rank = star_wars.columns[9:15]

print(rank)
```

```python
mean_rank = star_wars[rank].mean()
print(mean_rank)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

mean_rank.plot.bar()
plt.show()

# ranking_5 is the most favourite film for the franchise.
```

```python
sum_seen = star_wars[seen].sum()
print(sum_seen)
```

```python
sum_seen.plot.bar()
plt.show()
```

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
males_sumrank = males[rank].sum()

print(males_sumrank)

males_sumrank.plot.bar()
```

```python
males_sumseen = males[seen].sum()

print(males_sumseen)

males_sumseen.plot.bar()
```

```python
print(females[rank].sum())
print(females[seen].sum())
```

Interestingly, more males watches episodes 1-3, but males liked them far less than females did.
